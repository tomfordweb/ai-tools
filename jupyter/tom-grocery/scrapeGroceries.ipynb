```python
import scan_receipt
items = scan_receipt.scan({"directory": "data/mperks-t"})
for item in items:
    print(item)
```

```python
import sys
from langchain_ollama import ChatOllama
from langchain.messages import AIMessage
from langchain_core.prompts import PromptTemplate
from langchain.messages import HumanMessage, AIMessage, SystemMessage


llm = ChatOllama(
    model="tinyllama",
    temperature=0,
    base_url="http://ollama:11434"
)

conversation = [
    SystemMessage("You are an expert in grocery store inventory and products sold in grocery stores. You are responsible for reading this list of upc and product abbreviations. Do not make any modifications to the data. Return only rows from the list that are food. Remove drinks and energy drinks."),
    HumanMessage('\n'.join(items)),
]

response = llm.invoke(conversation)
print(response.content)
```
