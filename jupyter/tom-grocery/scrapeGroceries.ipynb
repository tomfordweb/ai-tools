```python
import scan_receipt

items = scan_receipt.scan({"directory": "data/mperks-t"})
for item in items:
    print(item)
```

```python
import sys
from langchain_ollama import ChatOllama
from langchain.messages import AIMessage
from langchain_core.prompts import PromptTemplate
from langchain.messages import HumanMessage, AIMessage, SystemMessage


llm = ChatOllama(
    model="gemma3:270m",
    temperature=0,
    base_url="http://ollama:11434"
)

conversation = [
    SystemMessage("Review the submitted list consisting of UPC's and product names seen on a receipt. Filter out any candy or drink items from the list"),
    HumanMessage('\n'.join(items)),
]

response = llm.invoke(conversation)
print(response.content)
```
