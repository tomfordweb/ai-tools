```python
    # Initialize the ChatOpenAI model, pointing to your LM Studio server
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(
        model="liquid/lfm2-1.2b",
        temperature=0.0,  # Or your desired temperature
        base_url="http://127.0.0.1:1234/v1",
        api_key="not-needed"
    )
```

```python
response = llm.invoke("what is the meaning of life as a software enginner")
print(response.content)
```
